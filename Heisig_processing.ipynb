```python
import pandas as pd
df = pd.read_csv('data/heisig.csv')
df
```

```python
# Creates dictionary of Simplified character
heisig_rsh_dict = {}

for index, row in df.iterrows():
    character_info = {}
    
    character = row['SH']
    if character == character:
        try:
            int(row['RSH #'])
            number = int(row['RSH #'])
        except:
            number = None
            
        character_info['Number'] = number
        character_info['Traditional'] = row['TH']
        character_info['Kanji'] = row['K']
        #print(row['RSH Keyword'])
        character_info['Keyword'] = row['RSH Keyword']
        heisig_rsh_dict[character] = character_info
```

```python
# Saves dictionaries to pickle
import pickle

with open('data/heisig_rsh_dict.pkl', 'wb') as handle:
    pickle.dump(heisig_rsh_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
